```python
import pandas as pd
```

```python
df_k1 = pd.read_table(
    'moscow_kinosat_1.txt', 
    sep='\s+',
    names=[
        'Access', 
        'Start Day',
        'Start Month',
        'Start Year',
        'Start Time',
        'End Day',
        'End Month',
        'End Year',
        'End Time',
        'Duration'
    ]
)

df_k1.drop(columns=['Access'], inplace=True)

# Concatenate date and time columns and convert to datetime 
df_k1['Start Datetime'] = pd.to_datetime(df_k1['Start Day'].astype(str) + ' ' + df_k1['Start Month'] + ' ' + df_k1['Start Year'].astype(str) + ' ' + df_k1['Start Time'])
df_k1['End Datetime'] = pd.to_datetime(df_k1['End Day'].astype(str) + ' ' + df_k1['End Month'] + ' ' + df_k1['End Year'].astype(str) + ' ' + df_k1['End Time'])

df_k1.drop(
    columns=[
        'Start Day',
        'Start Month',
        'Start Year',
        'Start Time',
        'End Day',
        'End Month',
        'End Year',
        'End Time',
    ], 
    inplace=True
)
df_k1['Setellite'] = 'k1'
```

```python
df_k2 = pd.read_table(
    'moscow_kinosat_2.txt', 
    sep='\s+',
    names=[
        'Access', 
        'Start Day',
        'Start Month',
        'Start Year',
        'Start Time',
        'End Day',
        'End Month',
        'End Year',
        'End Time',
        'Duration'
    ]
)

df_k2.drop(columns=['Access'], inplace=True)

df_k2['Start Datetime'] = pd.to_datetime(df_k2['Start Day'].astype(str) + ' ' + df_k2['Start Month'] + ' ' + df_k2['Start Year'].astype(str) + ' ' + df_k2['Start Time'])
df_k2['End Datetime'] = pd.to_datetime(df_k2['End Day'].astype(str) + ' ' + df_k2['End Month'] + ' ' + df_k2['End Year'].astype(str) + ' ' + df_k2['End Time'])

df_k2.drop(
    columns=[
        'Start Day',
        'Start Month',
        'Start Year',
        'Start Time',
        'End Day',
        'End Month',
        'End Year',
        'End Time',
    ], 
    inplace=True
)
df_k2['Setellite'] = 'k2'
```

```python
t1 = []
t2 = []
for indx, el in df_k1.iterrows():
    t1.append((el['Start Datetime'], el['End Datetime'], el['Setellite']))

for indx, el in df_k2.iterrows():
    t2.append((el['Start Datetime'], el['End Datetime'], el['Setellite']))
```

```python
# t1 = [(1, 4, 'k1'), (6, 8, 'k1'), (10, 11, 'k1')]
# t2 = [(2, 5, 'k2'), (9, 15, 'k2')]

timeline = []
for el in t1 + t2:
    timeline.append( (el[0], el[2]) )
    timeline.append( (el[1], el[2]) )
```

```python
timeline = sorted(timeline)
```

```python
timeline
```

```python
# Алгоритм разбиения пересекающихся отрезков
new_table = {}
is_opened = {}

for el in timeline:
    if el[1] not in new_table.keys():
        new_table[el[1]] = [[el[0], None]]
        is_opened[el[1]] = True
        
    opened = [i for i in is_opened.items() if i[1]]
    if not is_opened[el[1]]:
        is_opened[el[1]] = True
        new_table[el[1]].append([el[0], None])

    for opened_el in opened:
        if new_table[opened_el[0]][-1][0] != el[0]:
            new_table[opened_el[0]][-1][1] = el[0]
            if opened_el[0] != el[1]:
                new_table[opened_el[0]].append([el[0], None])
            else:
                is_opened[el[1]] = False
```

```python
new_table
```

```python
res = []
for k in new_table.keys():
    for v in new_table[k]:
        res.append((k, tuple(v), v[0]))
df = pd.DataFrame.from_records(res)
df = df.sort_values(by=[2]).reset_index()

mutex = []
for uv in df[1].unique():
    tmp = df[df[1] == uv].index.values
    if len(tmp) > 1:
        mutex.append(tmp)
```
